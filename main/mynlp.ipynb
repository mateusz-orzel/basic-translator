```python
import pandas as pd
import numpy as np
import torch
import tensorflow as tf
import nltk
nltk.download('punkt')
```

```python
eng = []
pol = []

with open("../static/text files/eng.txt", "r", encoding="utf-8") as f:
    eng = f.readlines()

with open("../static/text files/pol.txt", "r", encoding="utf-8") as f:
    pol = f.readlines()
```

# Tokenization and Cleaning

```python
import string

pol_clean = []
eng_clean = []

for pol_sent, eng_sent in zip(pol, eng):
    pol_sent = pol_sent.translate(str.maketrans('', '', string.punctuation)).lower()
    eng_sent = eng_sent.translate(str.maketrans('', '', string.punctuation)).lower()
    
    pol_tokens = nltk.word_tokenize(pol_sent)
    eng_tokens = nltk.word_tokenize(eng_sent)
    
    pol_clean.append(pol_tokens)
    eng_clean.append(eng_tokens)
```

```python
all_sentences = pol_clean + eng_clean

# Create a frequency distribution of the words in the merged list of sentences
all_words = [word for sent in all_sentences for word in sent]
word_freq = nltk.FreqDist(all_words)

# Create a vocabulary of the most frequent words in the corpus
vocab_size = 10000
min_freq = 5
vocab_words = [word for (word, freq) in word_freq.most_common(vocab_size) if freq >= min_freq]

# Create dictionaries that map words to integers and vice versa
word2id = {"<PAD>": 0, "<UNK>": 1}
id2word = {0: "<PAD>", 1: "<UNK>"}

for i, word in enumerate(vocab_words):
    word2id[word] = i + 2
    id2word[i + 2] = word
```
